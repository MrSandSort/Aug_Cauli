#### Import packages for Augmentation (CNN)

```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array, array_to_img
```

```python
train_datagen= ImageDataGenerator(rescale= 1.0/255, rotation_range= 40, 
                                 width_shift_range=0.2, height_shift_range=0.2,
                                 shear_range=0.2, zoom_range=0.2, horizontal_flip=True,
                                 fill_mode='nearest')
```

```python
img= load_img('Cauli_01.jpeg')
```

```python
img
```

```python
x= img_to_array(img)
```

```python
x = x.reshape((1,) + x.shape)
```

```python
i=0

for batch in train_datagen.flow(x, batch_size=1, save_to_dir= 'preview', save_prefix='cauli', save_format='jpeg'):
    i+=1
    if i>20:
        break
```
